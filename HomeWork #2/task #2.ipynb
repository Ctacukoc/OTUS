### Проверяем при помощи визуализации некоторые гипотезы

**1-ая Гипотеза**: Чем больше у подростка свободного времени, тем больше алкоголя он употреляет

```python
df["all_alc"] = df["Dalc"] + df["Walc"] # Общий показатель употрление алкоголя в неделю (вне зависимости от дней недели)
```

```python
gr_df = df.groupby(["freetime"], as_index = False).agg({"all_alc":"mean"}, ax=ax2)
g = sns.factorplot(x="freetime", y="all_alc", data=gr_df)
g.set_ylabels("Ср. оценка употр. алк. (2 - мало, 10 - много)")
g.set_xlabels("Свободное время (1 - очень мало, 5 - очень много)")
```

**Вывод**: Гипотеза подтвердилась

** 2-ая Гипотеза **: Чем ниже образование у родителей, тем больше алкоголя употребляет подросток

```python
gr_df = df.groupby(["Medu"], as_index = False).agg({"all_alc":"mean"}, ax=ax2)
g = sns.factorplot(x="Medu", y="all_alc", data=gr_df)
g.set_ylabels("Ср. оценка употр. алк. (2 - мало, 10 - много)")
g.set_xlabels("Уровень образование матери (0-нет, 1-до 4 класса, 2-с 5 по 9 класс, 3 - среднее образ. 4 - высшее обр.)")
```

```python
gr_df = df.groupby(["Fedu"], as_index = False).agg({"all_alc":"mean"})
g = sns.factorplot(x="Fedu", y="all_alc", data=gr_df)
g.set_ylabels("Ср. оценка употр. алк. (2 - мало, 10 - много)")
g.set_xlabels("Уровень образование  (0-нет, 1-до 4 класса, 2-с 5 по 9 класс, 3 - среднее образ. 4 - высшее обр.)")
```

Вывод: По уровню образованию отца не все так однозначно, что касается матери - гипотеза подтверждается, за исключением случая (2 - образование с 1-4 классы), что может быть связано с малым количеством респондентов с этой характеристикой. В случае с отцом, гипотеза не подтвержадется. 

**3 - ая гипотеза:** Если у подростка отсутствует дома интернет, то он будет употреблять больше алкоголя.

```python
gr_df = df.groupby(["internet"], as_index = False).agg({"all_alc":"mean"})
g = sns.factorplot(x="internet", y="all_alc", data=gr_df)
g.set_ylabels("Ср. оценка употр. алк. (2 - мало, 10 - много)")
g.set_xlabels("Наличие дома интернета (y - да, n - нет)")
```

Вывод: в целом гипотеза подтвердилась, скорей всего при отсутствии интернета дома, подростки больше всего времени проводят на улице. 
Однако отметим, что средняя оценка употребления алкоголя выше всего на 0.12

**4-ая гипотеза:** Если у подростка родители живут раздельно или разведены, то он употреляет больше алкоголя

```python
gr_df = df.groupby(["Pstatus","sex"], as_index = False).agg({"all_alc":"mean"})
g = sns.factorplot(x="Pstatus", y="all_alc", hue="sex", kind="bar", data=gr_df)
g.set_ylabels("Ср. оценка употр. алк. (2 - мало, 10 - много)")
g.set_xlabels("Родители (T - живут вместе, А - живут раздельно)")
```

Вывод: Девушки-подростки употреляет одинаковое кол-во алкоголя, вне зависимости от семейного положения своих родитедей, 
а юноши употребляют больше алкоголя, если если родители живут раздельно.
